```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for Stock column
plt.figure(figsize=(8, 5))
sns.histplot(df['Stock'], bins=20, kde=True)
plt.title("Distribution of Stock")
plt.show()
```

```python
# Scatter plot between Stock and TotalStock
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Stock', y='TotalStock', data=df)
plt.title("Stock vs Total Stock")
plt.show()

# Correlation heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(df[['Stock', 'CommodityStock', 'TotalStock']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data (replace 'your_file.csv' with your actual file path)
df = pd.read_csv(r"C:\Users\chaud\OneDrive\Desktop\project\Cases (Oldest) 1970 (1).csv")

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Drop rows with NaT in 'Date' column if any
df = df.dropna(subset=['Date'])

# Extract the year from the 'Date' column
df['Year'] = df['Date'].dt.year

# Group by 'Year' and calculate the sum of 'TotalStock' for each year
yearly_stock = df.groupby('Year')['TotalStock'].sum().reset_index()

# Plotting the year-wise total stock trend
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='TotalStock', data=yearly_stock, marker='o')
plt.title("Year-wise Total Stock Trend")
plt.xlabel("Year")
plt.ylabel("Total Stock")
plt.grid()
plt.show()
```
